```python
# UN ABB es como un arbol genealogico
# pero con NÚMEROS
# 1. Cada nodo tiene máximo 2 hijos
# 2. Izquierda: numeros menores
# 3. Derecha: numeros mayores

"""
             10
            /  \
           5    15
          / \   / \
         3   7 12  20
"""
```

```python
# Leccion 1: creando nodo
# Un nodo es como una caja
# guarda: el valor (número)
# guarda: dos flechas apuntando a sus hijos

class Nodo:
    def __init__(self, valor):
        self.valor = valor
        self.izq = None # Hijo izquierdo (numeros menores)
        self.der = None # Hijo derecho (numeros mayores)

# Creamos un nodo simple
raiz = Nodo(10)
print(f"Nodo creado con valor: {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izq}")
print(f"Hijo derecho: {raiz.der}")
```

```python
# Desafio 1: Crea un nodo con el valor 50
# y muestra su valor.

raiz = Nodo(50)
print(f"Nodo creado con valor: {raiz.valor}")
```

```python
# Conenctando Nodos
# Vamos a crear un arbol pequeñito
"""
             10
            /  \
           5    15
"""

# Crear raíz
raiz = Nodo(10)
raiz.izq = Nodo(5)
raiz.der = Nodo(15)

print(f"Raiz: {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izq.valor}")
print(f"Hijo derecho: {raiz.der.valor}")
```

```python
# Desafío 2:
# Crea un arbol con raiz 20
# Hijo izq 10, hijo der 30
# imprime el resultado

"""
             20
            /  \
           10  30
"""
# Crear raíz
raiz = Nodo(20)
raiz.izq = Nodo(10)
raiz.der = Nodo(30)

print(f"Raiz: {raiz.valor}")
print(f"Hijo izquierdo: {raiz.izq.valor}")
print(f"Hijo derecho: {raiz.der.valor}")
```

```python
# Desafio 3
# Regla:
# Si el valor es menor -> ir a la izquierda
# Si el valor es mayor -> ir a la derecha
# Repetir hasta encontrar un espacio vacio


def insertar(raiz, valor):
    if raiz is None:
      return Nodo(valor)
      # Si el valor es menor, ir a la izq
    if valor < raiz.valor:
      raiz.izq = insertar(raiz.izq, valor)
      # Si el valor es mayor, ir a la der
    else:
      raiz.der = insertar(raiz.der, valor)

    return raiz

# Ejemplo construir este arbol:

"""
             10
            /  \
           5    15
          /
         3
"""

raiz = None
raiz = insertar(raiz, 10)
raiz = insertar(raiz, 5)
raiz = insertar(raiz, 15)
raiz = insertar(raiz, 3)

print(f"Raiz: {raiz.valor}")
print(f"Izquierda: {raiz.izq.valor}")
print(f"Derecha: {raiz.der.valor}")
print(f"Izquierda de Izquierda: {raiz.izq.izq.valor}")
```
